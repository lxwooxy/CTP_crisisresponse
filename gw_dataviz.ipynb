# Function to build dictionaries for column names

```python
service_dictionary_path = '../311_dictionary.csv'
nypd_dictionary_path = '../nypd_dictionary.csv'

def build_dictionary(dictionary_path):
    with open(dictionary_path, 'r') as file:
        dictionary = {}
        for line in file:
            key = line.strip().split(',')[0]
            #remove \ufeff character, BOM
            key = key.lstrip('\ufeff')
            # value is the concatenation of the rest of the line
            value = ','.join(line.strip().split(',')[1:])
            dictionary[key] = value
    return dictionary
```

```python
service_dict = build_dictionary(service_dictionary_path)
nypd_dict = build_dictionary(nypd_dictionary_path)
```

```python
print("Keys in service_dict:")
print(service_dict.keys())

print("Keys in nypd_dict:")
print(nypd_dict.keys())
```
